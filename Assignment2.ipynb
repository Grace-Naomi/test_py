```python
username = "admin"
password = "secret123"

if not username:
    print("Error: Username is required.")
elif not password:
    print("Error: Password is required.")
elif username != "admin":
    print("Error: Incorrect username.")
elif password != "secret123":
    print("Error: Incorrect password.")
else:
    print("Login successful!")
```

```python
bill_amount = 85.00
tip_rate = 0.18
people = 4

tip_amount = bill_amount * tip_rate
total_amount = bill_amount + tip_amount
amount_per_person = total_amount / people

print("Bill amount: $", bill_amount)
print("Tip amount: $", round(tip_amount, 2))
print("Total amount: $", round(total_amount, 2))
print("Amount per person: $", round(amount_per_person, 2))
```

```python
hours = 2
minutes = 30
seconds = 45

# Formula
total_seconds = (hours * 3600) + (minutes * 60) + seconds

print("Total seconds:", total_seconds)
```
