# List slicing

```python
numbers = [10, 20, 30, 40, 50, 60, 70]
numbers[4:7]
```

```python
numbers = [10, 20, 30, 40, 50, 60, 70]

print(numbers[0:3])    # [10, 20, 30]       — first 3 items
print(numbers[2:5])    # [30, 40, 50]       — items at index 2, 3, 4
print(numbers[:3])     # [10, 20, 30]       — from start (0 is default)
print(numbers[4:])     # [50, 60, 70]       — from index 4 to end
print(numbers[:])      # [10, 20, 30, 40, 50, 60, 70]  — full copy
print(numbers[::2])
```

# iTERATE THRUGH THE LISTS

```python
team = ["Rohit", "Virat", "Dhoni", "Bumrah", "Jadeja"]

for i in team:
    print(i)
```

```python
team = ["Rohit", "Virat", "Dhoni","Rohit", "Bumrah", "Jadeja"]

for i in team:
    if team.count(i) > 1:
        team.remove(i)

print(team)
```

```python
for i, v in enumerate(team):
    print(i, v)
```
