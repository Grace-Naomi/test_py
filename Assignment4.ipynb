```python
# Define the prime checking function
def is_prime_number():
    num = 17
    if num < 2:
        print(f"{num} is not a prime number")
        return
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print(f"{num} is not a prime number")
            return
    print(f"{num} is a prime number")

# Call the function
is_prime_number()
```
