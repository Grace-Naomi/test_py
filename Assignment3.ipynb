```python
# Input list
names = ["Alice", "Bob","Emily", "David", "Oliver", "Sarah"]

# Vowels
vowels = "AEIOU"

# Counter
count = 0

# Check each name
for name in names:
    if name[0].upper() in vowels:
        count += 1

# Output
print("Names starting with vowels:", count)
```

```python
# Input list
numbers = [45, 23, 67, 89, 12, 56]

max_value = numbers[0]
position = 0

# Use enumerate to get index and value
for index, value in enumerate(numbers):
    if value > max_value:
        max_value = value
        position = index

# Output
print("Maximum value:", max_value)
print("Position:", position)
```

```python
# Input list and target value
numbers = [5, 2, 8, 2, 9, 2, 1, 2]
target = 2

# List to store positions
positions = []

# Loop through the list to find all occurrences
for index, value in enumerate(numbers):
    if value == target:
        positions.append(index)

# Output
print(f"Value {target} found at positions: {positions}")
```
